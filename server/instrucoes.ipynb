# **Instruções funcionamento da API**

### **1 - Conexão com o Oracle DB**

Na função *conexao_db()* você precisa passar suas credenciais do banco de dados Oracle. Para maior segurança:

1- crie um arquivo *credenciais.json* na na mesma pasta do projeto *server.py*

2- Adicione suas credenciais

ex:

![Captura de tela 2024-05-16 160048.png](<attachment:Captura de tela 2024-05-16 160048.png>)


Com isso você irá conectar ao seu banco de dados Oracle.




### **2 - Criando as tabelas no banco**

É necessário que você tenha a tabela criada no seu banco com produtos dentro dela.

Para isso, você pode usar o arquivo *criar_tabelas.sql* que está na pasta *scripts* do projeto.



__1-__ Copie o Create Table e cole no Sql developer



__2-__ Clique no botão play, no canto superior esquerdo do script



__3-__ Logo após você precisará clicar no botão para commitar as alterações



__4-__ Você pode conferir se os produtos foram adicionados, executando a query *SELECT*, contida na pasta scripts

![Design sem nome (13).png](<attachment:Design sem nome (13).png>)



### **3 - Ligando o servidor e retornando a o conteúdo da API**

Agora com o banco de dados criado e com dados nele, podemos fazer as requições no banco e gerar um arquivo JSON, via API, para ser consumido no endpoint


__1-__ Entrar no arquivo *server.py*

__2-__ Abra o terminal e escreva ***python server.py***
    
    
Isso inicializará o servidor Flask no navegador

__3-__ Acessar o navegador e produrar por ***http://localhost:8080/api/produtos***


__4-__ Ao acessar o navegador, deverá ser retornado um arquivo JSON, com todos os produtos cadastrados no seu banco de dados
